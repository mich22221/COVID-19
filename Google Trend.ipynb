```
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()

```

```
pytrend.build_payload(kw_list = ["Economist"])
df = pytrend.interest_over_time()
```

```
df
```

```
df.reset_index().plot(x="date", y="Economist", figsize=(120, 10), kind ="line")
```
